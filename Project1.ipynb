# Play a round

```python
class Subgame:
    
    def __init__(self, board, pit): # used for minimax and alphabeta players 
    
        self.player1_pits = [1, 2, 3, 4, 5, 6]
        self.player2_pits = [8, 9, 10, 11, 12, 13]
        
        self.connected = {1: 13, 2:12, 3: 11, 4: 10, 5: 9, 6: 8,
                          13:1, 12: 2, 11: 3, 10: 4, 9: 5, 8: 6}
                          
        self.stones_held = 0   
        
        self.board = board #current state becomes board 
        self.current_pit = pit #action entered becomes current pit 
        
        if self.current_pit < 7:#checks which player is currently playing the round
            self.player = 1#
        else:#
            self.player = 2 #
        
        self.ans = False #changes if the player is able to go again
        
        self.go() # calls go to start the round 
        
    def go(self):
        
        self.pick_up(self.current_pit) #picks ups stones from current pit 
        self.drop(self.current_pit + 1) # drops to next pit and starts recursion
        self.check_over() #checks if the game is over 
    
    def pick_up(self, pit):

        self.stones_held += self.board.get(pit) #holds stones in current pit
        self.board[pit] = 0 #after holding these stones there are no stones left in pit
        
    def drop(self, pit):
        
        self.board[pit] += 1 #adds 1 stone to current pit 
        self.stones_held -= 1 #takes away stone from stones held 
        
        if self.stones_held == 0: #after there are no stones left in hand
        
            if pit == 14 or pit == 7: #if able to go again changes the ans to True (last stone lands in respective store)
                self.ans = True 
            
            elif self.board[pit] == 1 and pit != 7 and pit != 14 and self.board.get(self.connected.get(pit)) != 0: #if last stone is isolated, check to see if player can capture stones
                self.capture(pit)
            
        else:    
       
            if self.player == 1 and pit == 13: #player 1 cannot drop in player 2 store
                pit = 1 #skips 14 
                self.drop(pit) #continues recursion from pit 1
                
            elif self.player == 2 and pit == 6: #player 2 cannot drop in player 1 store 
                pit = 8 # skips 7
                self.drop(pit) #continues from pit 8
            
            elif pit == 14: # connects 14 to 1 
                pit = 1
                self.drop(pit) #continues recurion from pit 1
            
            else:
                pit += 1 #continue recursion if none of the conditions are met 
                self.drop(pit)
                
    def capture(self, pit):

        if self.player == 1 and pit in self.player1_pits: # if player 1 is able to capture 
            self.pick_up(self.connected.get(pit)) #picks up from vertically connected pit 
            self.pick_up(pit) #picks up from current pit 
            self.board[7] += self.stones_held #adds the stones held to the player 1 store 
            self.stones_held = 0 # not holding stones anymore

        elif self.player == 2 and pit in self.player2_pits: #if player 2 can capture 
            self.pick_up(self.connected.get(pit)) #picks up from connected and current pit 
            self.pick_up(pit)
            self.board[14] += self.stones_held #adds the stones to the player 2 store 
            self.stones_held = 0 #not holding stones anymore 
                
        else: 
            pass
        
    def check_over(self):
        
        if self.board.get(1) == 0 and self.board.get(2) == 0 and self.board.get(3) == 0 and self.board.get(4) == 0 and self.board.get(5) == 0 and self.board.get(6) == 0:
           
            for i in range(8, 14): # if there are no stones on player 1 side of the board pick up all stones on player 2 side 
                self.pick_up(i)
            self.board[14] += self.stones_held #add them to their store 
            self.stones_held = 0 #no more stones held
            
        elif self.board.get(8) == 0 and self.board.get(9) == 0 and self.board.get(10) == 0 and self.board.get(11) == 0 and self.board.get(12) == 0 and self.board.get(13) == 0:
            
            for i in range(1, 7): #if there are no stones on player 2 side, pick up all stones on player 1 side 
                self.pick_up(i)
            self.board[7] += self.stones_held #adds the stones to their store 
            self.stones_held = 0 #not holding anymore stones 
        
        else:
            pass 
    
    def return_board(self): #returns board 
        return self.board


    def go_again(self): #returns if able to go again
        return self.ans
 
```

```python
class Human:
    
    def __init__(self): 
        self.title = 'human' #just states name 
        self.state = None
        
    def make_move(self, options_list, state): #available_pits is a list
        
        
        self.state = state #reads in the board 
        self.opts = options_list #reads in what options the player has 
        self.legal = []
        
        for i in self.opts:
            if self.state[i] != 0:
                self.legal.append(i) # if there is a pit with no stones in it then it is not added to the legal moves list   
        
        action = input('enter pit to pick up from: ') # enters an action 
        while action.isdigit() != True or int(action) not in self.legal: # if the action is not legal then you must continue to enter until it is
            action = input('Invalid pit, please enter again: ') 
              
        return action # returns choice 
    
    def declare(self):
        return self.title 
```

```python
class Random_move:
    
    def __init__(self): 
        self.title = 'random'
        self.state = None
        
    def make_move(self, options_list, state):
        
        self.state = state
        self.opts = options_list
        self.legal = []
        
        for i in self.opts:
            if self.state[i] != 0: #checks for legal moves 
                self.legal.append(i)
        
        action = choice(self.legal) # chooses a random choice in the legal moves list
        
        return action #returns action
    
    def declare(self):
        return self.title
```

```python
class Minimax:
    
    def __init__(self): # initializes all properties of the algortihm
        
        self.state = None
        self.title = 'Minimax'
        
        self.maxplayer_pits = None
        self.minplayer_pits = None
        self.maxplayer_store = None
        self.minplayer_store = None
        
        self.depth = 5 #expansion cannot go beyond 5
        self.decision = 0 #initial decision is not a legal move and changes when make_move is ran
        
    def make_move(self, opts_list, state):
        
        self.state = state #reads in the state of the board 
        self.maxplayer_pits = opts_list #reads in which player and their pits
        
        if self.maxplayer_pits == [1, 2, 3, 4, 5, 6]: #if maxplayer is player 1 
            self.minplayer_pits = [8, 9, 10, 11, 12, 13] #minplayer is player 2
            self.maxplayer_store = 7 #respective stores 
            self.minplayer_store = 14
            
        elif self.maxplayer_pits == [8, 9, 10, 11, 12, 13]: #same for player 2
            self.minplayer_pits = [1, 2, 3, 4, 5, 6]
            self.maxplayer_store = 14
            self.minplayer_store = 7
            
        self.max_move(self.state, self.maxplayer_pits, self.depth) #calls max_move to start recursive stage with the maxplayer's pits and initial limit

        return self.decision #self.decision is returned after the recurion is concluded 
    
    def max_move(self, state, actions, limit):
        
        v = -1_000_000 
        
        for i in actions:

            if state[i] != 0: #checks which of the pits are legal moves 
                
                x = Subgame(state.copy(), i)#plays round with the state of the board and the action chosen
                
                if x.go_again():
                    s = self.max_move(x.return_board(), self.maxplayer_pits.copy(), limit) #if the player is able to go again then then it will recall maxmove function with same limit 
                
                elif (state[7] + state[14] == 48) or limit == 0: #if limit is reached or game has reached an end it will return the utility of the chain 
                    s = state[self.maxplayer_store] - state[self.minplayer_store] #takes the difference to eventually find best outcome for maxplayer
                
                else:
                    s = self.min_move(x.return_board(), self.minplayer_pits.copy(), limit - 1) #if neither of the above options calls minmove with a shorter shorter depth
                
                if v < s: # if the value returned is larger than the current v value 
                    v = s # sets new value 
                    if limit == self.depth: #if the limit is equal to the initial depth(top of list)
                        self.decision = i #sets the decision that has the highest value 
                        
        return v

        
    def min_move(self, state, actions, limit):
        
        v = 1_000_000
            
        for i in actions:
            
            if state[i] != 0: #for each action that is legal 
                x = Subgame(state.copy(), i) #play round
                
                if x.go_again(): #if able to move again, do so with same limit
                    s = self.min_move(x.return_board(), self.minplayer_pits.copy(), limit)
                    
                elif (state[7] + state[14] == 48) or limit == 0: #if end of game or limit reached return value 
                    s = state[self.maxplayer_store] - state[self.minplayer_store]     
                    
                else:
                    s = self.max_move(x.return_board(), self.maxplayer_pits.copy(), limit - 1) #if neither of the above conditions call maxmove
                
                if v > s:
                    v = s  #if value returned is less than current value switch them        
                
        return v
    
    def declare(self):
        return self.title #used in game class for viewing the board 
            
```

```python
class Alphabeta:
    
    def __init__(self): #initializes all values of the class 
        
        self.state = None
        self.title = 'AlphaBeta'
        
        self.maxplayer_pits = None
        self.minplayer_pits = None
        self.maxplayer_store = None
        self.minplayer_store = None
        
        self.depth = 5 #limiter of 5
        self.decision = 0 #decision to be returned once recursion has concluded 
        
    def make_move(self, opts_list, state):
        
        self.state = state #reads in current state of the board 
        self.maxplayer_pits = opts_list #reads in which player is the max player 
        
        if self.maxplayer_pits == [1, 2, 3, 4, 5, 6]: #if player 1 then sets stores and pits to appropriate players 
            self.minplayer_pits = [8, 9, 10, 11, 12, 13]
            self.maxplayer_store = 7
            self.minplayer_store = 14
            
        elif self.maxplayer_pits == [8, 9, 10, 11, 12, 13]:#same for player 2
            self.minplayer_pits = [1, 2, 3, 4, 5, 6]
            self.maxplayer_store = 14
            self.minplayer_store = 7
            
        self.max_move(self.state, self.maxplayer_pits, self.depth, -1_000_000, 1_000_0000) #calls max_move with initialized state, available actions and initial alpha and beta values to be changed throughout the recursion

        return self.decision #returns final choice of best value
    
    def max_move(self, state, actions, limit, alpha, beta):
        
        v = -1_000_000
        
        for i in actions:

            if state[i] != 0: #finds legal moves 
                
                x = Subgame(state.copy(), i)#plays round with every available action
                
                if x.go_again():
                    s = self.max_move(x.return_board(), self.maxplayer_pits, limit, alpha, beta) #if able to go again, do so with the same limit, alpha and beta values
                
                elif (state[7] + state[14] == 48) or limit == 0: # if end of game or limit is reached return utility of that choice
                    s = state[self.maxplayer_store] - state[self.minplayer_store]
                
                else:
                    s = self.min_move(x.return_board(), self.minplayer_pits, limit - 1, alpha, beta) # if neither of the above conditions then call minmove with 1 less limit and same alpha/beta values 
                
                if v < s: #if returned value is more than the current value then change current value 
                    v = s
                    if limit == self.depth: #returns one of the original legal moves 
                        self.decision = i
                
                if v >= beta: #returns value is greater than beta then returns move to prune expansion
                    return v
                
                alpha = max(alpha, v) #sets the alpha value after first expansion
        return v # returns utility

        
    def min_move(self, state, actions, limit, alpha, beta):
        
        v = 1_000_000
            
        for i in actions:
            
            if state[i] != 0: #looks through available moves 
                x = Subgame(state.copy(), i)   # plays round
                
                if x.go_again():
                    s = self.min_move(x.return_board(), self.minplayer_pits, limit, alpha, beta) #goes again if able to 
                    
                elif (state[7] + state[14] == 48) or limit == 0: # returns if end of game or limit reached 
                    s = state[self.maxplayer_store] - state[self.minplayer_store]    
                    
                else:
                    s = self.max_move(x.return_board(), self.maxplayer_pits, limit - 1, alpha, beta) # if neiter, calls maxmove with 1 less depth and alpha beta values 
                
                if v > s: # if v more than returned utility, changes vurrent value 
                    v = s    
                    
                if v <= alpha: # if value is less than alpha, prune expansion
                    return v
                    
                beta = min(beta, v) #sets the beta value after first expansion
                
        return v
    
    def declare(self):
        return self.title #returns the title for sake of board view
            
```

```python
#game 
class Kalahgame:
    
    def __init__(self, player1 = input('enter player 1: '), player2 = input('enter player 2: ')):  
        
        self.player1 = player1 #the players are entered by the user(s) that are playing 
        self.player2 = player2
        
        self.player_list = ['human', 'random', 'minimax', 'alphabeta', 'most']
        
        while self.player1 not in self.player_list: 
            self.player1 = input('Invalid player. Enter player 1 again: ')
        
        if self.player1 == 'human':  #assigns the player to the respective input to with classes above 
            self.player1 = Human()
        elif self.player1 == 'random':
            self.player1 = Random_move()
        elif self.player1 == 'minimax':
            self.player1 = Minimax()        
        elif self.player1 == 'alphabeta':
            self.player1 = Alphabeta()       
        elif self.player1 == 'most':
            self.player1 = Most_stones()
            
        while self.player2 not in self.player_list:
            self.player2 = input('Invalid player. Enter player 2 again: ')
                  
        if self.player2 == 'human':
            self.player2 = Human()
        elif self.player2 == 'random':
            self.player2 = Random_move()
        elif self.player2 == 'minimax':
            self.player2 = Minimax()         
        elif self.player2 == 'alphabeta':
            self.player2 = Alphabeta()      
        elif self.player2 == 'most':
            self.player2 = Most_stones()
        
        
        self.player1_pits = [1, 2, 3, 4, 5, 6]
        self.player2_pits = [8, 9, 10, 11, 12, 13]
        
        self.board = {1: 4, 2: 4, 3: 4, 4: 4, 5: 4, 6: 4, 7: 0,
                      8: 4, 9: 4, 10: 4, 11: 4, 12: 4, 13: 4, 14: 0} # initial state of the board 
        
        self.connected = {1: 13, 2:12, 3: 11, 4: 10, 5: 9, 6:8,
                          13:1, 12: 2, 11: 3, 10: 4, 9: 5, 8: 6} #establishing the connections between pits for capturing pieces
        
        self.stones_held = 0   #number stones in hand is initially 0
        self.current_pit = 0  # initial pit is changed when the player has chosen a move 
        self.player1_choice = ' '
        self.player2_choice = ' '
        self.player_turn = randint(1, 2) #randomly decides who is going first 
        self.start_game() #starts game 
        
        
    def start_game(self):
        
        self.print_board() #initial state of the game 
        self.turn_start() #starts the player who goes first 
    
    def turn_start(self):
        
        if self.player_turn%2 == 1: # if the player turn is odd then it is player 1s turn 
            print('\nplayer 1 turn \n')
            
            self.player1_choice = self.player1.make_move(self.player1_pits, self.board) # has the player select a move 
            self.current_pit = int(self.player1_choice) #the selected move then becomes the current pit 
            if (self.player1.title == 'AlphaBeta' or self.player1.title == 'Minimax') and self.board[self.current_pit] == 0:
                for i in range(1, 7):
                    if self.board[i] != 0:
                        self.current_pit = i
            
            
        else:
            print('\nplayer 2 turn \n') # if it is not odd then it is player 2s turn 

            self.player2_choice = self.player2.make_move(self.player2_pits, self.board)
            self.current_pit = int(self.player2_choice)
            
            if (self.player2.title == 'AlphaBeta' or self.player2.title == 'Minimax') and self.board[self.current_pit] == 0:
                for i in range(8, 14):
                    if self.board[i] != 0:
                        self.current_pit = i
                    
            
        print(self.current_pit)
        
        self.pick_up(self.current_pit) # picks up stones from the current pit 
        self.current_pit += 1 #moves the current pit to the next to proceed to drop action
        self.drop(self.current_pit) #proceeds to drop procedure 
        self.turn_end() #concludes the turn after out of stones 

    def turn_end(self):
        
        self.print_board() #shows the state of the game 
        self.scoring()#shows the scores of the respective players 
        self.current_pit = 0 #resets the pit 
        
        print('Player 1 chose: ' + str(self.player1_choice)) #shows what the player chose 
        print('Player 2 chose: ' + str(self.player2_choice))
        
        self.player1_choice = ' '
        self.player2_choice = ' '
        
        if self.check_over(): # checks if the game is over 
            self.someone_won() # if returned true then it checks who won and how many points each player received 
            self.print_board()
            self.scoring()
            
        else:
            self.player_turn += 1 #if the game isn't over then the turn goes to the next player and starts the round over 
            self.turn_start()
        
    
    def scoring(self):
        
        print('player 1 has: ' + str(self.board.get(7)) + 
              '\n' + 'player 2 has: ' + str(self.board.get(14))) #shows the score of respective players 
    
    def check_over(self):
        
        result = None # eventually want to return whehter the game is over or not
        
        if self.board.get(1) == 0 and self.board.get(2) == 0 and self.board.get(3) == 0 and self.board.get(4) == 0 and self.board.get(5) == 0 and self.board.get(6) == 0:
           #if all the pits on the player 1s side of the board are empty then it will pick up all the pieces of the other side of the board and drop them in their respective store
            for i in range(8, 14):
                self.pick_up(i) # picks up all stones in each player 2 pit 
            self.board[14] += self.stones_held #adds the stones to player 2 store 
            self.stones_held = 0 #resets the number of stones held
            
            result = True # returns that the game is over 
            
        elif self.board.get(8) == 0 and self.board.get(9) == 0 and self.board.get(10) == 0 and self.board.get(11) == 0 and self.board.get(12) == 0 and self.board.get(13) == 0:
            # does the same for player 2, bring all stones on player 1 side and drop them in their store 
            for i in range(1, 7):
                self.pick_up(i)
            self.board[7] += self.stones_held
            self.stones_held = 0
            
            result = True #returns true if the game is over 
        
        else:
            result = False # if the game is not over it will return that the game is not over 
        
        return result
    
    def someone_won(self):
        
        if self.board.get(7) > self.board.get(14):
            print('\nPlayer 1 Wins!') # after it is established that the game is over, it shows which player 1 based on who has more in their store 

        elif self.board.get(14) > self.board.get(7):
            print('\nPlayer 2 Wins!')
            
        else:
            print('\nDraw.')
    
    def print_board(self): # prints the board 
        
        s = '        ' 
        s2 = '        '
        
        for i in [13, 12, 11, 10, 9, 8]:
            
            if self.board.get(i) < 10: # adjusts format based on the number of stones in the pit for player 2
                s += ' (' + str(self.board.get(i)) + ' ) ' # single digit number of stones 
            else:
                s+= ' (' + str(self.board.get(i)) + ') ' #double digit of stones 
                
        for j in range(1, 7):
            
            if self.board.get(j) < 10:
                s2 += ' (' + str(self.board.get(j)) + ' ) ' #adjusts format for player 1 
            else:
                s2 += ' (' + str(self.board.get(j)) + ') '
        print('\n')
        print('       -------------------------------------')
        print('       | 13  | 12  | 11  | 10  |  9  |  8  |')
        print('       -------------------------------------')
        print('                                               Player 2: ' + self.player2.declare()) #shows what type of player 2 
        print(s) #prints stones in respective pits 
        print('(' + str(self.board.get(14)) + ')   --------------------------------------   (' + str(self.board.get(7)) + ')')
        print(s2) # prints stones in respective pits 
        print('                                               Player 1: ' + self.player1.declare()) #shows type of player 1 
        print('       -------------------------------------')
        print('       |  1  |  2  |  3  |  4  |  5  |  6  |')
        print('       ------------------------------------- \n')
        

    def pick_up(self, pit):

        self.stones_held += self.board.get(pit) #picks up all the stones in the pit after the pit has been selected 
        self.board[pit] = 0 # establishs that there are no stones left in the pit 
    
    def drop(self, pit):
        
        self.board[pit] += 1 #drops one stone in the pit 
        self.stones_held -= 1 #takes away stone from hand 
        
        if self.stones_held == 0: # after there are no stones in the hand 
        
            if pit == 14 or pit == 7: # if the last stone is dropped in the store (only the respective player can drop in their store)
                self.player_turn -= 1 # subtracts from player turn so they can go again
            
            elif self.board[pit] == 1 and pit != 7 and pit != 14 and self.board.get(self.connected.get(pit)) != 0:
                self.capture(pit) #if there is only one stone in the pit and its not a store then it will call capture function to steal stones from other player based on connection dictionary 
                 # cannot capture if there are no stones in the connected pit 
        else:    # if there are still stones in the hand 
       
            if self.player_turn%2 == 1 and pit == 13: # if player 1 and they are in pit 13 
                pit = 1 #skips 14 and goes to 1 
                self.drop(pit) #drops in pit 1
                
            elif self.player_turn%2 == 0 and pit == 6: #if player 2 and they are in pit 6
                pit = 8 # skips 7 and goes to pit 8
                self.drop(pit) #drops in 8
            
            elif pit == 14: #connects pit 14 to pit 1 
                pit = 1
                self.drop(pit)
            
            else:
                pit += 1 # if normal drop sequence goes to next pit and drops 
                self.drop(pit)
    
    def capture(self, pit): # captures stones 

        if self.player_turn%2 == 1 and pit in self.player1_pits: # if player 1 and the last stone ends up being the only stone in that pit on player 1 side of the board 
            self.pick_up(self.connected.get(pit)) # picks up stones from connected pit based on connected dictionary 
            self.pick_up(pit) #picks up isolated stone 
            self.board[7] += self.stones_held # adds them to the store 
            self.stones_held = 0 #resets number of stones in hand 

        elif self.player_turn%2 == 0 and pit in self.player2_pits: #if player 2 and last stone is isolated on player 2 side of the board 
            self.pick_up(self.connected.get(pit)) # picks up from connected pit 
            self.pick_up(pit) # picks up from current pit 
            self.board[14] += self.stones_held  #adds them to the store 
            self.stones_held = 0 #resets hand 
                
        else: 
            pass
Kalahgame() #starts game 
```
